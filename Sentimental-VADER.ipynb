```python
def sentiments(rating):
    if (rating == 5.0) or (rating == 4.0):
        return "Positive"
    elif rating == 3.0:
        return "Neutral"
    elif (rating == 2.0) or (rating == 1.0):
        return "Negative"
# Add sentiments to the data
tra_data["Sentiment"] = tra_data["overall"].apply(sentiments)
tra_data["Sentiment"][:30]
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
test_polarity = np.array(tra_data['Sentiment'])
for review in tra_data:
    print('Review:-')
    print(tra_data['reviewText'])
    print('Actual Labeled Sentiment:', test_polarity)
    print()
    
```

```python
def analyze_sentiment_vader_lexicon(review, threshold = 0.3, verbose = False):
    scores = analyzer.polarity_scores(review)  
    if scores['compound'] > threshold:
        binary_sentiment = 'Positive'  
    elif scores['compound'] == threshold:
        binary_sentiment = 'Neutral'
    else: binary_sentiment ='Negative'
    if verbose:                             
        print('VADER Polarity (Binary):', binary_sentiment)
        print('VADER Score:', round(scores['compound'], 2))
    return binary_sentiment,scores['compound']  
```

```python
VADER_polarity = [analyze_sentiment_vader_lexicon(review, threshold=0.3) for review in tra_data['reviewText']]
VADER_polarity_df = pd.DataFrame(VADER_polarity, columns = ['VADER Polarity','VADER Score'])
VADER_polarity_test_df.head(20)
```
